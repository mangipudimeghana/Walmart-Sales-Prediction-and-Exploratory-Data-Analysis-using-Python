# Question 'a': If the weekly sales are affected by the unemployment rate, if yes - which stores are suffering the most?

```python
## Scatter plot on Impact of Unemployment on Sales

sns.scatterplot(x=df.Unemployment, y=df.Weekly_Sales,hue=df.Store, s=80)
plt.xlabel('Unemployment', fontsize=10, labelpad=20)
plt.ylabel('Sales', fontsize=10, labelpad=20)
plt.title('Impact of Unemployment on Sales')
plt.show()
```

# Answer to Question 'a':

A higher unemployment index would generally result in a dip in sales as individuals tend to decrease overall spending.

From our scatter plot, it is easier to conclude the following information:

â€¢ For the given stores, there seems to be a visible decrease in sales when the unemployment index is higher than 11

â€¢ There seems to be a significant drop in sales for stores mentioned in scatter plot, when the
unemployment index increases

â€¢ Highest recorded sales for stores mentioned occur around the unemployment index of 8 to 10; 
this gives ambiguous ideas about the impact of unemployment on sales for each of the store

# Question 'b':If the weekly sales show a seasonal trend, when and what could be the reason?

```python
## Plot on Weekly_Sales vs Date

df.plot(x='Date', y='Weekly_Sales',figsize=(20,10),fontsize=20)
plt.xlabel('Date', fontsize=10, labelpad=20)
plt.ylabel('Weekly_Sales', fontsize=10, labelpad=20)
plt.title('Weekly_Sales vs Date')
plt.show()
```

# Answer to Question 'b'

From the above plot we can say that weekly_sales shows a seasonal trend. 

The highest sales seem to occur in the late spring and early summer (around May and June). This could be due to a variety of factors such as Seasonal changes in consumer behavior, Seasonal variations in product availability, Holidays and events

The lowest sales appear to occur in the winter months (around December and January). This could be due to Decreased consumer spending during the holiday season, Seasonal changes in consumer behavior, Fewer business days

# Question 'c': Does temperature affect the weekly sales in any manner?

```python
## Scatter plot on Impact of Temperature on Weekly_Sales

sns.scatterplot(x=df.Temperature, y=df.Weekly_Sales,hue=df.Store, s=40,c='red')
plt.xlabel('Temperature', fontsize=10, labelpad=20)
plt.ylabel('Sales', fontsize=10, labelpad=20)
plt.title('Impact of Temperature on Sales')
plt.show()
```

# Answer to Question 'c'

From the above scatter plot we can conclude that:

The highest sales occur for most stores between the range of 40 to 80 degrees Fahrenheit, thus proving the idea that pleasant weather encourages
higher sales. 

Sales are relatively lower for very low and very high temperatures but seem to be adequately high for favorable climate conditions.


# Question 'd': How is the Consumer Price index affecting the weekly sales of various stores?

```python
## Scatter plot on Impact of CPI(Consumer Price Index) on Weekly_Sales

sns.scatterplot(x=df.CPI, y=df.Weekly_Sales,hue=df.Store, s=40,c='red')
plt.xlabel('CPI', fontsize=10, labelpad=20)
plt.ylabel('Weekly_sales', fontsize=10, labelpad=20)
plt.title('Impact of CPI on Sales')
plt.show()
```

# Answer to Question 'd'

CPI (Consumer Price Index) is defined as the measure of the average change over time in the prices paid by urban consumers for a market basket of consumer goods and services

In our scatter plot above, we can identify three different clusters around different ranges of CPI, while there seems to be no visible relationship between the change in
CPI and weekly sales for different stores (sales still occur at high CPI rates)

```python
# Checking for correlation for all columns

df.corr()
```

```python
# Plotting HeatMap 

fig=plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap="rainbow")
```

```python
# Importing variance_inflation_factor and preparing list of columns to which VIF needs to be performed

from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in df.columns:
    if((df[col].dtype!='object') & (col!='Weekly_Sales')):
        col_list.append(col)
```

```python
col_list
```

```python
# Checking for VIF value for all the columns in col_list

X = df[col_list] 
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF values'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
vif_data
```

```python
# Dropping Fuel_Price column as it is having high VIF(26.73)

X=X.drop(['Fuel_Price'], axis = 1)
```

```python
# Again checking VIF for remaining columns

vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF values'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
vif_data
```

```python
# Dropping Unemployment column as it is having high VIF(12.75)

X=X.drop(['Unemployment'], axis = 1)
```

```python
# Again checking VIF for remaining columns

vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF values'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
vif_data
```

```python
# Dropping Temperature column as it is having high VIF(10.06)

X=X.drop(['Temperature'], axis = 1)
```

```python
# Again checking VIF for remaining columns

vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF values'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
vif_data
```

```python
# Considering only required columns for Predictions

df=df.loc[:,["Store","Holiday_Flag","CPI","Weekly_Sales"]]
```

```python
df
```

```python
# Fitting data into X and Y

x=df.iloc[:,:-1]
y=df["Weekly_Sales"]
```

```python
x
```

```python
y
```

```python
# Importing train_test_split and splitting the data into training and testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.7, random_state=34)
```

```python
x_train
```

```python
# Importing Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
```

```python
# Creating model

r_model = RandomForestRegressor()
```

```python
# Training model

r_model.fit(x_train,y_train)
```

```python
# checking for the predictions

predictions=r_model.predict(x_test)
```

```python
predictions
```

```python
# Importing all from sklearn

from sklearn.metrics import *
```

```python
# Checking the R2 score for our model 

r2_score(predictions,y_test)  
```

# FORECASTING THE SALES OF WALMART STORES

```python
# scatter plot between observed and predicted values of weekly sales from Random forest regressor

plt.figure(figsize = (12,6))
plt.title("MODEL : RANDOM FOREST")
plt.xlabel("ORIGINAL VALUES")
plt.ylabel("PREDICTED VALUES")
sns.scatterplot(x = y_test, y = predictions, color='red')
plt.show()
```

```python
# Importing Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor
```

```python
# Creating model

d_model = DecisionTreeRegressor()
```

```python
# checking for the predictions

predictions_d=d_model.predict(x_test)
```

```python
# Importing all from sklearn

from sklearn.metrics import *
```
